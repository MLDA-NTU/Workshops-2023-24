## For Loop

```python
sum=0                                 # we create a variable called sum and assign it to a value of 0
for i in range(2):                    # you can replace i with a, b or cat dog etc, range is exclusive and starts from 0
    print("i is:", i)                 # press tab to indent if needed.
    print("sum is:", sum)
    sum=sum+1
                                      # i will be destroyed at the end of the for loop
```

```python
print(i)
```

## While Loop

```python
i=0                                   # create a variable i and assign it to a value of 0
sum=0                                 # sum would be 2 from for loop above so we reset it to 0
while i<2:
    print("i is:", i)
    print("sum is:", sum)
    sum=sum+1
    i=i+1
    print("end of iteration")
print("end of while loop")
```

## Conditionals

```python
for i in range(10):
    if i % 2 == 0: # % - modulus operator -> returns the remainder after division
        print(f"{i} is even") # f-strings (a method): or alternatively, use .format(i)
    else:
        print(f"{i} is odd")
```

```python
a=True                                 # a is a boolean variable assigned to true
i=0                                    # variable i is assigned to a value of 0

while(a):                              # while true means the loop continues, while false means it will exit the loop
    if i>=3:                           # conditional operator, > is more than, < is less than, >= is more than or equal to, <=
        print("escaping")              # when i>=3, it prints escaping
        a=False                        # changes a to false, will finish the remaining part in 11,12 before leaving loop
    elif i!=2:                         # checks for not equal
        print("i is not equal 2")
    elif i==2:                         # == checks for equal
        print("i is equal 2")
    print("i is :", i)                  # outside the if, elif, it will run as per normal
    i=i+1                              # last line of while loop
print("escaped while loop")
```

# Familiarity with Object Oriented Programming (OOP)

A **programming paradigm (or model)**

On the concept of "objects" which are reusable, self-contained pieces of code representing real-world entities or concepts.

```python
# Syntax of classes:

class Student:

    university = "NTU" # class attribute

    def __init__(self, name, age): # instance attributes
        self.name =  name
        self.age = age

    def study(self, subject):
        self.subject = subject
        print(f"{self.name} is now studying {self.subject}")

a = Student("Ava", 18)
b = Student("Ben", 17)
print(a.name)
a.study("Math")
```

```python
x = np.array([2, 4, 6]) # create rank 1 array
A = np.array([[1, 3, 5], [2, 4, 6]]) # create rank 2 array
B = np.array([[1, 2, 3], [4, 5, 6]])

print("Matrix A: \n")
print(A)

print("\nMatrix B: \n")
print(B)
```

```python
# Indexing/Slicing

print(A[0, :]) # index the first "row" and all columns
print(A[1, 2]) # index the second row, third column entry
print(A[:, 1]) # index entire second column
```

```python
# Arithmetic Examples

C = A * 2 # multiply every element of A by two
D = A * B # elementwise multiplication rather than matrix multiplication
E = np.transpose(B)
F = np.matmul(A, E) # performs matrix multiplication -- could also use np.dot()
G = np.matmul(A, x) # performs matrix-vector multiplication -- again could also use np.dot()

print("\n Matrix E (the transpose of B): \n")
print(E)

print("\n Matrix F (result of matrix multiplication A x E): \n")
print(F)

print("\n Matrix G (result of matrix-vector multiplication A*x): \n")
print(G)
```

```python
# max operation examples

X = np.array([[3, 9, 4], [10, 2, 7], [5, 11, 8]])
all_max = np.max(X) # gets the maximum value of matrix X
column_max = np.max(X, axis=0) # gets the maximum in each column -- returns a rank-1 array [10, 11, 8]
row_max = np.max(X, axis=1) # gets the maximum in each row -- returns a rank-1 array [9, 10, 11]

# In addition to max, can similarly do min. Numpy also has argmax to return indices of maximal values
column_argmax = np.argmax(X, axis=0) # note that the "index" here is actually the row the maximum occurs for each column

print("Matrix X: \n")
print(X)
print("\n Maximum value in X: \n")
print(all_max)
print("\n Column-wise max of X: \n")
print(column_max)
print("\n Indices of column max: \n")
print(column_argmax)
print("\n Row-wise max of X: \n")
print(row_max)
```

```python
# Sum operation examples
# These work similarly to the max operations -- use the axis argument to denote if summing over rows or columns


total_sum = np.sum(X)
column_sum = np.sum(X, axis=0)
row_sum = np.sum(X, axis=1)

print("Matrix X: \n")
print(X)
print("\n Sum over all elements of X: \n")
print(total_sum)
print("\n Column-wise sum of X: \n")
print(column_sum)
print("\n Row-wise sum of X: \n")
print(row_sum)
```
