```python
import torch     # main pytorch
import torch.nn as nn # components build neural net
import torch.functional as F  # Function
import torch.optim as optim # Optimizer
```

```python
# Drop out
layer = nn.Dropout(p=0.5) # probability = 0.9
input = torch.tensor([1.0 ,2.0 ,3.0 ],dtype=torch.float)
output = layer(input)
print(input)
print(output)
```

```python
class Model(nn.Module): # Class
  def __init__(self, lr):
    super().__init__()
    self.layer1 = nn.Linear(128,128)  # Initialize
    self.layer2 = nn.Linear(128, 10)
    self.learning_rate = lr
    # Storage space of variable

  def forward(self,x): # input. FORWARD is a CONVENTION defined by the library class
    x = nn.ReLU(self.layer1(x))
    x = self.layer2(x) #Call
    return x

  def forward_prop(self,x):
    return x


# model = Model() #initialize
# model(x) # Calling the model
# model.forward_prop(x)
```

```python
model = Model()
param = model.parameters  # no bracket, variable
print(param)



param_call = model.parameters() # a different generator object
print(param_call)
```

```python
# Optimizer
sgd_optimizer = optim.SGD(model.parameters(), lr = 0.01) # model.parameters() goes here, lr = learning rate, eta (n like symbol)
adam_optimizer = optim.Adam(model.parameters(), weight_decay = 0.1)


scheduler = optim.lr_scheduler.StepLR(adam_optimizer, step_size = 30, gamma = 0.1) # After calling 30 times, muliply lr with gamma

for epoch in range(100):
  # input, label
  #adam_optimizer.zero_grad()
  #output = model(input)
  #loss = loss_function(output, label)
  #loss.backward()
  # Train here
  # Validate here
  adam_optimizer.step()
  scheduler.step() # 30 times learning rate decreases
  print("Epoch: ", epoch)

```
